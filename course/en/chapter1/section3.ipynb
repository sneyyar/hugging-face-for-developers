## Data collection

### Subtask:
Gather a dataset of English textual descriptions and corresponding Python Playwright code snippets. This dataset will be used to train the model.


**Reasoning**:
Searching for and potentially scraping data can be a time-consuming and complex process involving external resources. Manually creating a small dataset is a more controlled and feasible first step within this interactive environment to demonstrate the data format required for the task. This will also allow for a concrete starting point if further data collection is needed.



```python
import pandas as pd

# Manually create a small dataset of English descriptions and Playwright code
data = {
    'description': [
        'Navigate to the website "https://playwright.dev".',
        'Click on the button with the text "Get started".',
        'Type "Playwright" into the input field with the placeholder "Search docs".',
        'Close the current page.',
        'Go back to the previous page.'
    ],
    'playwright_code': [
        'page.goto("https://playwright.dev")',
        'page.click("text=Get started")',
        'page.fill("input[placeholder=\\"Search docs\\"]", "Playwright")',
        'page.close()',
        'page.go_back()'
    ]
}

df = pd.DataFrame(data)

# Display the created dataframe
display(df)

# Save the dataframe to a CSV file
df.to_csv('playwright_code_dataset.csv', index=False)
```

## Model training

### Subtask:
Train the selected model on the prepared dataset. This involves feeding the model pairs of text descriptions and code snippets so it learns to generate code from the descriptions.


**Reasoning**:
Load the T5 model and tokenizer, prepare the dataset by tokenizing the descriptions and code, define training arguments, create a Trainer, and start training.



```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForSeq2Seq
from datasets import Dataset
import torch

# 1. Load the t5-small model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

# 2. Prepare the dataset for training
# Convert the pandas DataFrame to a Hugging Face Dataset
dataset = Dataset.from_pandas(df)

def preprocess_function(examples):
    # Tokenize the input descriptions (English text)
    model_inputs = tokenizer(examples["description"], max_length=128, truncation=True, padding="max_length")

    # Tokenize the target code snippets (Playwright code)
    labels = tokenizer(examples["playwright_code"], max_length=128, truncation=True, padding="max_length")

    # Set the labels for the model, replacing padding token id with -100
    model_inputs["labels"] = labels["input_ids"]
    # Replace tokenizer.pad_token_id in labels with -100
    model_inputs["labels"] = [
        [(label if label != tokenizer.pad_token_id else -100) for label in labels_example] for labels_example in model_inputs["labels"]
    ]

    return model_inputs

# Apply the preprocessing function to the dataset
tokenized_dataset = dataset.map(preprocess_function, batched=True)

# 3. Define the training arguments
training_args = TrainingArguments(
    output_dir="./results",  # Output directory for checkpoints and logs
    num_train_epochs=3,      # Number of training epochs
    per_device_train_batch_size=2, # Batch size per device during training
    per_device_eval_batch_size=2,  # Batch size for evaluation
    warmup_steps=500,        # Number of warmup steps for learning rate scheduler
    weight_decay=0.01,       # Strength of weight decay
    logging_dir="./logs",    # Directory for storing logs
    logging_steps=10,
)

# 4. Create a DataCollator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

# 5. Create a Trainer object
trainer = Trainer(
    model=model,                         # The instantiated 🤗 Transformers model to be trained
    args=training_args,                  # Training arguments, defined above
    train_dataset=tokenized_dataset,     # Training dataset
    data_collator=data_collator,         # Data collator to handle padding and batching
)

# 6. Start the training process
trainer.train()
```

## Model Evaluation

### Subtask:
Evaluate the trained model's performance using appropriate metrics for code generation, such as BLEU or exact match.

**Reasoning**:
Generate Playwright code using the trained model for the descriptions in the dataset and calculate the BLEU score to evaluate the model's performance.

```python
from transformers import pipeline
from datasets import load_metric

# Create a pipeline for text generation using the fine-tuned model
# Since this is a sequence-to-sequence model, we use the text2text-generation pipeline
generator = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Generate code for the descriptions in the dataset
descriptions = df['description'].tolist()
generated_codes = generator(descriptions)

# Extract the generated text from the pipeline output
generated_code_texts = [item['generated_text'] for item in generated_codes]

# Get the reference code snippets from the dataset
reference_codes = df['playwright_code'].tolist()

# Load the BLEU metric
bleu_metric = load_metric("bleu")

# Prepare the references for the BLEU metric (it expects a list of lists)
# Each reference is a list of tokens (words)
reference_codes_tokenized = [[code.split()] for code in reference_codes]
generated_code_tokenized = [code.split() for code in generated_code_texts]


# Compute the BLEU score
results = bleu_metric.compute(predictions=generated_code_tokenized, references=reference_codes_tokenized)

print("Generated Code:")
for description, generated_code in zip(descriptions, generated_code_texts):
    print(f"Description: {description}")
    print(f"Generated Code: {generated_code}")
    print("-" * 20)

print(f"\nBLEU score: {results['bleu']}")
```
